```python
import re
import random
from pathlib import Path

import numpy as np
import pandas as pd
from PIL import Image, ImageDraw, ImageFont
import tifffile as tiff

# -------------------
# CONFIG
# -------------------
SAMPLE_DIR   = Path("/mnt/volumec/Aswath/MAIT STUDY/raw_data/image_files/LHCC35")  # folder with the TIFFs
SAMPLE_ID    = "LHCC35"                                       # used to filter CSV
CSV_PATH     = "selected_samples.csv"                         # columns: Tissue,XMin,YMin,XMax,YMax
OUT_DIR      = Path("out_vis")
SEED         = 42

# CSV column names
ID_COL = "Tissue"
XMIN, YMIN, XMAX, YMAX = "XMin", "YMin", "XMax", "YMax"

# Marker to overlay on the SAME Hoechst crop
OTHER_MARKER = "CD19"
PAD = 0  # crop padding (pixels) around bbox
DRAW_BOX_ON_COMPOSITE = True
BOX_COLOR = (255, 0, 0)
BOX_WIDTH = 3

# -------------------
# Helpers
# -------------------
def find_hoechst_tif(folder: Path) -> Path:
    cands = list(folder.glob("*.tif")) + list(folder.glob("*.tiff"))
    for p in cands:
        if re.search(r"(hoechst|dapi)", p.name, re.IGNORECASE):
            return p
    if cands:
        return max(cands, key=lambda p: p.stat().st_size)
    raise FileNotFoundError(f"No TIFFs found in {folder}")

def find_marker_tif(folder: Path, marker_name: str) -> Path:
    for p in list(folder.glob("*.tif")) + list(folder.glob("*.tiff")):
        if re.search(marker_name, p.name, re.IGNORECASE):
            return p
    raise FileNotFoundError(f"No TIFF found for marker '{marker_name}' in {folder}")

def _read_plane(path: str) -> np.ndarray:
    try:
        img = tiff.imread(path, key=0)
    except Exception:
        with tiff.TiffFile(path) as tf:
            try:
                img = tf.asarray(key=0, maxworkers=1)
            except Exception:
                img = tf.pages[0].asarray()
    img = np.squeeze(np.asarray(img))
    if img.ndim > 2:
        img = img.reshape(img.shape[-2], img.shape[-1])
    if img.ndim != 2:
        raise ValueError(f"Expected 2D plane, got shape {img.shape} from {path}")
    return img

def to_uint8_contrast(img: np.ndarray) -> np.ndarray:
    img = img.astype(np.float32)
    lo, hi = np.percentile(img, [1, 99])
    if hi <= lo:
        hi = lo + 1.0
    img = np.clip((img - lo) / (hi - lo), 0, 1)
    return (img * 255).astype(np.uint8)

def pil_from_gray(gray: np.ndarray) -> Image.Image:
    gray8 = to_uint8_contrast(gray)
    rgb = np.stack([gray8]*3, axis=-1)
    return Image.fromarray(rgb)

def draw_box(img: Image.Image, bbox, color=(255, 0, 0), width=3, label=None):
    x0, y0, x1, y1 = map(int, bbox)
    draw = ImageDraw.Draw(img)
    for t in range(width):
        draw.rectangle([x0+t, y0+t, x1-t, y1-t], outline=color)
    if label:
        font = ImageFont.load_default()
        tw, th = draw.textbbox((0, 0), label, font=font)[2:]
        pad = 0
        bx0, by0 = x0, max(0, y0 - th - 2*pad)
        draw.rectangle([bx0, by0, bx0 + tw + 2*pad, by0 + th + 2*pad], fill=(255,255,255))
        draw.text((bx0 + pad, by0 + pad), label, fill=color, font=font)

def save_crop(img: Image.Image, bbox, pad=40, out_path=OUT_DIR/"crop.png"):
    x0, y0, x1, y1 = map(int, bbox)
    W, H = img.size
    x0p, y0p = max(0, x0 - pad), max(0, y0 - pad)
    x1p, y1p = min(W, x1 + pad), min(H, y1 + pad)
    img.crop((x0p, y0p, x1p, y1p)).save(out_path)

def norm_01(im: np.ndarray) -> np.ndarray:
    im = im.astype(np.float32)
    lo, hi = np.percentile(im, [1, 99])
    if hi <= lo:
        hi = lo + 1.0
    return np.clip((im - lo) / (hi - lo), 0, 1)

def hoechst_marker_composite_on_same_crop(
    hoechst_arr: np.ndarray,
    marker_arr:  np.ndarray,
    bbox,
    pad=10,
    draw_box_on_composite=True,
    box_color=(255,0,0),
    box_width=1,
) -> Image.Image:
    H, W = hoechst_arr.shape
    x0, y0, x1, y1 = map(int, bbox)
    x0p, y0p = max(0, x0 - pad), max(0, y0 - pad)
    x1p, y1p = min(W, x1 + pad), min(H, y1 + pad)

    h_crop = hoechst_arr[y0p:y1p, x0p:x1p]
    m_crop = marker_arr[y0p:y1p, x0p:x1p]

    h_norm = norm_01(h_crop)
    m_norm = norm_01(m_crop)

    rgb = np.zeros((h_norm.shape[0], h_norm.shape[1], 3), dtype=np.float32)
    rgb[..., 2] = h_norm
    rgb[..., 1] = m_norm
    img = Image.fromarray((rgb * 255).astype(np.uint8))

    if draw_box_on_composite:
        bx0, by0 = x0 - x0p, y0 - y0p
        bx1, by1 = x1 - x0p, y1 - y0p
        draw = ImageDraw.Draw(img)
        for t in range(box_width):
            draw.rectangle([bx0+t, by0+t, bx1-t, by1-t], outline=box_color)

    return img

# -------------------
# Main
# -------------------
if __name__ == "__main__":
    OUT_DIR.mkdir(parents=True, exist_ok=True)

    df = pd.read_csv(CSV_PATH)
    need = {ID_COL, XMIN, YMIN, XMAX, YMAX}
    missing = need - set(df.columns)
    if missing:
        raise ValueError(f"CSV missing columns: {missing}")

    hoechst_path = find_hoechst_tif(SAMPLE_DIR)
    arr = _read_plane(str(hoechst_path))
    img = pil_from_gray(arr)
    H, W = arr.shape

    sub = df[df[ID_COL] == SAMPLE_ID].copy()
    if sub.empty:
        sub = df.copy()

    random.seed(SEED)

    # =============== ORIGINAL FUNCTIONALITY ===============
    rand = sub.sample(n=1, random_state=SEED).iloc[0]
    bbox = (int(rand[XMIN]), int(rand[YMIN]), int(rand[XMAX]), int(rand[YMAX]))

    # Clip to bounds
    x0 = max(0, min(W-1, int(bbox[0])))
    y0 = max(0, min(H-1, int(bbox[1])))
    x1 = max(0, min(W-1, int(bbox[2])))
    y1 = max(0, min(H-1, int(bbox[3])))
    bbox = (x0, y0, x1, y1)

    overlay = img.copy()
    label = f"{SAMPLE_ID} | [{bbox[0]},{bbox[1]}â€“{bbox[2]},{bbox[3]}]"
    draw_box(overlay, bbox, color=BOX_COLOR, width=BOX_WIDTH, label=label)
    overlay.save(OUT_DIR / f"{SAMPLE_ID}_hoechst_random_cell_overlay.png")

    save_crop(overlay, bbox, pad=PAD, out_path=OUT_DIR / f"{SAMPLE_ID}_hoechst_random_cell_crop.png")

    marker_path = find_marker_tif(SAMPLE_DIR, OTHER_MARKER)
    marker_arr = _read_plane(str(marker_path))

    comp = hoechst_marker_composite_on_same_crop(
        arr, marker_arr, bbox,
        pad=PAD,
        draw_box_on_composite=DRAW_BOX_ON_COMPOSITE,
        box_color=BOX_COLOR,
        box_width=BOX_WIDTH,
    )
    comp.save(OUT_DIR / f"{SAMPLE_ID}_Hoechst_{OTHER_MARKER}_crop_overlay.png")

    # =============== NEW: THREE 2Ã— ENLARGED NUCLEUS PATCHES ===============
    def doubled_crop_with_lines(arr, bbox):
        x0, y0, x1, y1 = bbox
        nuc_w = x1 - x0
        nuc_h = y1 - y0

        # nucleus center (for one X line + one Y line)
        cx = x0 + nuc_w // 2
        cy = y0 + nuc_h // 2

        # doubled crop dimensions
        patch_w = nuc_w * 2
        patch_h = nuc_h * 2

        # compute enlarged crop boundaries
        x0p = max(0, cx - patch_w // 2)
        x1p = min(arr.shape[1], cx + patch_w // 2)
        y0p = max(0, cy - patch_h // 2)
        y1p = min(arr.shape[0], cy + patch_h // 2)

        # crop patch
        crop = arr[y0p:y1p, x0p:x1p]
        pil = pil_from_gray(crop)
        draw = ImageDraw.Draw(pil)

        # convert nucleus center to crop coordinates
        rel_cx = cx - x0p
        rel_cy = cy - y0p

        # === draw single vertical and horizontal lines ===
        # vertical (x-center)
        draw.line([(rel_cx, 0), (rel_cx, crop.shape[0])], fill=(255, 0, 0), width=2)

        # horizontal (y-center)
        draw.line([(0, rel_cy), (crop.shape[1], rel_cy)], fill=(0, 255, 0), width=2)

        return pil
```
